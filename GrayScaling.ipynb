# 1. Using cv2.cvtColor() function to convert to grayscale

> Note: OpenCV loads images in order of BGR instead of RGB -> needs to convert to RGB if we want to view the RGB colours in the original image.

```python
import cv2
print(cv2.__version__)
from matplotlib import pyplot as plt

image = cv2.imread('test_images/IMG_3167.PNG')

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # convert to RGB
# cv2.imshow('Original Image: ', image) # shows read image
# cv2.waitKey(0)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # converts to gray scale with cvtColor()

# cv2.imshow('Grayscaled Image: ', gray_image)
# cv2.waitKey(0)

# Binarisation of the image - 127 is the threshold value; > 127 outputs 255 (white), <= 127 outputs 0 (black) 
ret, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY) # converts to binary image 

# cv2.destroyAllWindows()

plt.subplot(1, 3, 1)
plt.title("Original Image: ")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Grayscaled Image: ")
plt.imshow(gray_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Binarised Image: ")
plt.imshow(binary_image, cmap='gray')
plt.axis("off")

plt.show()
```
